```python
from sklearn import tree
import pandas as pd
from sklearn.model_selection import train_test_split ,GridSearchCV

data = pd.read_csv("datadata.csv")
```

```python
from sklearn import preprocessing
from collections import defaultdict

d = defaultdict(preprocessing.LabelEncoder)

le = preprocessing.LabelEncoder()
# data = data.apply(le.fit_transform)
data.head()
fit = data.apply(lambda x: d[x.name].fit_transform(x))
X= fit[['userId', 'movieId','director','year','genre']]
y= fit[['rating']]

# fit = fit.apply(lambda x: d[x.name].inverse_transform(x))
# fit.head()

#make testng and training data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.tree import export_graphviz
from subprocess import call
from IPython.display import Image

tree_clf = tree.DecisionTreeClassifier(criterion = 'entropy')
tree_clf.fit(X_train, y_train)

export_graphviz(
        tree_clf,
        feature_names = X.columns,
        rounded=True,
        filled=True,
        out_file="project_tree.dot")

call(['dot','-Tpng', 'project_tree.dot', '-o', 'project_tree.png'], shell = True)
Image(filename='project_tree.png')
```
